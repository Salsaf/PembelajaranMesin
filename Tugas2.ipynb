```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt

# Theme: Favorite Pizza Toppings
toppings = ('Pepperoni', 'Mushroom', 'Onion', 'Peppers', 'Sausage')
votes = (15, 9, 6, 10, 8)
colors = ('#FF5733','#FFC300','#DAF7A6','#C70039','#900C3F')
explode = (0, 0, 0, 0.1, 0) # untuk menarik shadow
plt.title('Favorite Pizza Toppings')
plt.pie(
    votes,
    labels=toppings,
    autopct='%1.1f%%',
    explode=explode,
    shadow=True,
    colors=colors
)
plt.show()
```

Hasil visual dari pie chart yang menampilkan data tentang topping pizza favorit. Setiap bagian dari pie chart mewakili persentase dari suara yang diterima oleh setiap topping. Berikut adalah penjelasan visualnya:

-Bagian yang paling besar adalah topping  "Pepperoni", karena jumlah suaranya paling tinggi.
-Bagian selanjutnya topping "Peppers" karena jumlah suaranya cukup tinggi.
-Bagian lainnya akan mewakili topping-topping lainnya seperti "Mushroom", "Onion", dan "Sausage".
Topping yang memiliki efek explode (dalam hal ini "Peppers") akan sedikit terpisah dari lingkaran utama untuk menarik perhatian.

Warna yang digunakan untuk setiap bagian akan sesuai dengan yang ditentukan dalam variabel colors, memberikan visualisasi yang menarik dan mudah dipahami.

Keseluruhan, pie chart akan memberikan gambaran visual tentang preferensi topping pizza berdasarkan suara yang diterima.
